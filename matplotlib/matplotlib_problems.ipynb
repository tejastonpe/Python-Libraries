3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
x = []
y = []

with open('test.txt', 'r') as file:
    for line in file:
        values = line.split()
        x.append(int(values[0]))
        y.append(int(values[1]))

plt.plot(x, y, marker='o', linestyle='-', color='b', label="Data from file")

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot from test.txt')

plt.legend()
plt.show()
```

4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between
October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
03-10-16,774.25,776.065002,769.5,772.559998
04-10-16,776.030029,778.710022,772.890015,776.429993

06-10-16,779,780.47998,775.539978,776.859985

```python
df = pd.read_csv('fdata.csv')

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')

plt.figure(figsize=(10, 6))

plt.plot(df['Date'], df['Open'], marker='o', linestyle='-', label='Open')

plt.plot(df['Date'], df['High'], marker='o', linestyle='-', label='High')

plt.plot(df['Date'], df['Low'], marker='o', linestyle='-', label='Low')

plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', label='Close')

plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Alphabet Inc. Financial Data (October 3, 2016 to October 7, 2016)')

plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.plot(x, y, label='y = x^2', color='blue', marker='o')

current_xlim = plt.gca().get_xlim()
current_ylim = plt.gca().get_ylim()
print(f"Current X axis limits: {current_xlim}")
print(f"Current Y axis limits: {current_ylim}")

plt.xlim(-1, 6)
plt.ylim(-5, 30)

plt.title('Setting New Axis Limits')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.show()

new_xlim = plt.gca().get_xlim()
new_ylim = plt.gca().get_ylim()
print(f"New X axis limits: {new_xlim}")
print(f"New Y axis limits: {new_ylim}")
```

```python
11. Write a Python program to plot several lines with different format styles in one command
using arrays.
```

```python
x = np.linspace(0, 5, 10)
y1 = x ** 2
y2 = x ** 3
y3 = np.sqrt(x)

plt.plot(x, y1, '-o', x, y2, '--s', x, y3, '-.^')
plt.title('Multiple Lines with Different Format Styles')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend(['y = x^2', 'y = x^3', 'y = sqrt(x)'])
plt.show()
```

```python
12. Write a Python program to create multiple types of charts
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 35, 20, 15, 5]
#linechart
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
#barchart
plt.subplot(1, 2, 2)
plt.bar(categories, values, color='green')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python
import matplotlib.dates as mdates

dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

dates = [mdates.datestr2num(date) for date in dates]

plt.plot(dates, closing_values, marker='o', color='red', linestyle='--')

plt.grid(True, which='both', linestyle='-', linewidth=1)
plt.minorticks_on()
plt.xticks([])
plt.yticks([])

plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing Value of Alphabet Inc. (Oct 3, 2016 to Oct 7, 2016)')

plt.show()
```

15. Write a Python program to create multiple plots.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axs = plt.subplots(3, 1, figsize=(8, 6))  # 3 rows, 1 column

axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Function')

axs[1].plot(x, y2, color='green')
axs[1].set_title('Cosine Function')

axs[2].plot(x, y3, color='red')
axs[2].set_title('Tangent Function')

plt.tight_layout()
plt.show()
```
