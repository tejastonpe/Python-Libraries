1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0) 
x = np.random.rand(100) 
y = np.random.rand(100)  

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='skyblue')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Random Distribution')

plt.show()
```

2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other.

```python
np.random.seed(0)
x = np.random.rand(100)  
y = np.random.rand(100)  

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='skyblue', edgecolors='black', marker='o', s=100)  

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Empty Circles')

plt.show()
```

```python
3. Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes.
```

```python
np.random.seed(0)  
x = np.random.rand(50)  
y = np.random.rand(50)  
sizes = np.random.randint(10, 200, size=50)  


plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, color='skyblue', alpha=0.5)  

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Balls of Different Sizes')

plt.show()
```

5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.

```python
np.random.seed(0)
group1_height = np.random.normal(160, 10, 50)
group1_weight = np.random.normal(60, 5, 50)

group2_height = np.random.normal(170, 15, 50)
group2_weight = np.random.normal(70, 8, 50)

group3_height = np.random.normal(155, 8, 50)
group3_weight = np.random.normal(55, 4, 50)

plt.figure(figsize=(8, 6))
plt.scatter(group1_height, group1_weight, color='skyblue', label='Group 1')
plt.scatter(group2_height, group2_weight, color='orange', label='Group 2')
plt.scatter(group3_height, group3_weight, color='green', label='Group 3')

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot of Heights vs Weights for Three Groups')

plt.legend()
plt.show()
```
