1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.graph_objects as go
import numpy as np

x = np.random.rand(1000)
y = np.random.rand(1000)

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers'))
fig.update_layout(xaxis_title='X Coordinate', yaxis_title='Y Coordinate', title='Scatter Plot of Random 1000 X and Y Coordinates')

fig.show()
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates



```python

x = np.random.rand(100)
y = np.random.rand(100)

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='Line Plot'))
fig.update_layout(title='Line Plot for Random 100 Coordinates')
fig.show()

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', name='Scatter Plot'))
fig.update_layout(title='Scatter Plot for Random 100 Coordinates')
fig.show()
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it


```python
x = np.random.rand(500)
y = np.random.rand(500)

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', marker=dict(color='red', symbol='triangle-up', size=10), name='Styled Scatter Plot'))
fig.update_layout(title='Styled Scatter Plot for Random 500 Coordinates')
fig.show()
```
