1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Tips dataset from the URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
tips_data = pd.read_csv(url)

# Creating the swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(data=tips_data, x='size', y='total_bill', palette='Set2')

# Adding labels and title
plt.xlabel('Size of Party')
plt.ylabel('Total Bill ($)')
plt.title('Swarm Plot of Total Bill by Size of Party')

# Display the swarm plot
plt.show()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
tips_data = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.swarmplot(data=tips_data, x='day', y='total_bill', palette='muted')

plt.xlabel('Day')
plt.ylabel('Total Bill ($)')
plt.title('Swarm Plot of Total Bill against Day')

plt.show()
```
